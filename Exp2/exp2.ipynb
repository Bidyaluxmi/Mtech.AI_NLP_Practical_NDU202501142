```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer










```

```python
nltk.download('punkt') # downloads a pre-trained tokenizer model called “punkt”
nltk.download('punkt_tab')  # Extra tokenizer data (optional)
nltk.download('stopwords') # List of common words
nltk.download('wordnet')    # Dictionary for Lemmatization
nltk.download('omw-1.4')    # Multilingual WordNet support
```

```python
# Sample Text
text = """
Natural Language Processing is a fascinating field of Artificial Intelligence.
It enables computers to understand human language and process large amounts of text data efficiently.
"""

print("Original Text:\n", text)
```

```python
# ====================================================
# 1. TOKENIZATION
# ====================================================

tokens = word_tokenize(text)

print("\n----- Tokenization -----")
print(tokens)
```

```python
# ====================================================
# 2. STOP WORD REMOVAL
# ====================================================

stop_words = set(stopwords.words('english'))

filtered_words = [word for word in tokens if word.lower() not in stop_words]

print("\n----- After Stopword Removal -----")
print(filtered_words)
```

```python
# ====================================================
# 3. STEMMING
# ====================================================

stemmer = PorterStemmer()

stemmed_words = [stemmer.stem(word) for word in filtered_words]

print("\n----- After Stemming -----")
print(stemmed_words)
```

```python

# ====================================================
# 4. LEMMATIZATION
# ====================================================

lemmatizer = WordNetLemmatizer()

lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]

print("\n----- After Lemmatization -----")
print(lemmatized_words)
```
