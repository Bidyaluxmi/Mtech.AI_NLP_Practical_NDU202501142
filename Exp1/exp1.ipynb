```python
from transformers import pipeline
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
# =====================================================
# 1. SENTIMENT ANALYSIS
# =====================================================

def sentiment_analysis():
    print("\n===== SENTIMENT ANALYSIS =====")

    # Explicitly specify the default model to avoid the UserWarning
    sentiment_model = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")

    text = "Natural Language Processing is an amazing field of Artificial Intelligence."

    result = sentiment_model(text)

    print("Text:", text)
    print("Sentiment:", result)
```

```python
# =====================================================
# 2. TEXT SUMMARIZATION
# =====================================================
def text_summarization():
    print("\n===== TEXT SUMMARIZATION =====")

    from transformers import pipeline

    # Use stable model (works in Colab)
    summarizer = pipeline(
        "text2text-generation",
        model="sshleifer/distilbart-cnn-12-6"
    )

    article = """
    Natural Language Processing (NLP) is a subfield of artificial intelligence
    that focuses on enabling computers to understand and process human language.
    It combines linguistics, computer science, and machine learning to analyze text and speech.
    NLP applications include machine translation, chatbots, speech recognition,
    text summarization, sentiment analysis, and question answering.
    With the rise of deep learning models such as transformers,
    NLP systems have become more accurate and powerful.
    """

    summary = summarizer(article, max_length=50, min_length=20)

    print("Original Text:\n", article)
    print("\nSummary:\n", summary[0]['generated_text'])
```

```python
# 3. QUESTION ANSWERING
# =====================================================

def question_answering():
    print("\n===== QUESTION ANSWERING =====")

    qa_pipeline = pipeline("question-answering")

    context = """
    Natural Language Processing is a branch of Artificial Intelligence that
    enables computers to understand, interpret, and generate human language.
    NLP is used in applications like chatbots, machine translation,
    and voice assistants such as Siri and Alexa.
    """

    question = "What does NLP enable computers to do?"

    answer = qa_pipeline(question=question, context=context)

    print("Question:", question)
    print("Answer:", answer['answer'])
```

```python
# 4. INFORMATION RETRIEVAL (Semantic Search)
# =====================================================

def information_retrieval():
    print("\n===== INFORMATION RETRIEVAL =====")

    documents = [
        "Natural language processing enables computers to understand text.",
        "Deep learning is widely used in computer vision.",
        "Transformers are powerful models for NLP tasks.",
        "Machine learning is a subset of artificial intelligence."
    ]

    query = "What models are used in NLP?"

    model = SentenceTransformer('all-MiniLM-L6-v2')

    doc_embeddings = model.encode(documents)
    query_embedding = model.encode([query])

    similarity = cosine_similarity(query_embedding, doc_embeddings)

    best_match_index = np.argmax(similarity)

    print("Query:", query)
    print("Most Relevant Document:", documents[best_match_index])
```

```python
# MAIN PROGRAM
# =====================================================

if __name__ == "__main__":

    sentiment_analysis()

    text_summarization()

    question_answering()

    information_retrieval()
```
