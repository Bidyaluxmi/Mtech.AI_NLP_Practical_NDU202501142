```python
# pip install nltk spacy
# python -m spacy download en_core_web_sm

import nltk  # Natural Language Toolkit for basic NLP tasks
from nltk.stem import PorterStemmer, LancasterStemmer  # Import two stemming algorithms
import spacy  # Industrial-strength NLP library

# downloads a pre-trained tokenizer model called “punkt”
nltk.download('punkt')

# Load spaCy English model (contains vocabulary, rules, Lemmatizer,POS tagger, etc.)
# IMPORTANT: This will fail if 'en_core_web_sm' is not installed
nlp = spacy.load("en_core_web_sm")

# Initialize two different stemmers
porter = PorterStemmer()      # Less aggressive, commonly used
lancaster = LancasterStemmer()  # More aggressive, can distort words

# Sample words for testing
# Includes:
# - Inflectional forms (same root, different grammar)
# - Derivational forms (new words formed from root)
words = [
    "running", "runs", "ran",  # inflectional variations of "run"
    "better", "best",          # irregular forms of "good"
    "studies", "studying",     # verb variations
    "happiness", "happy",      # derivational
    "organization", "organize",# derivational
    "creation", "creative"     # derivational
]

# Print header for output formatting
print("=" * 70)
print("Morphological Analysis: Stemming vs Lemmatization")
print("=" * 70)

# Print column headers with alignment
print(f"{'Word':<15}{'Porter Stem':<20}{'Lancaster Stem':<20}{'Lemma':<15}")
print("-" * 70)

# Loop through each word
for word in words:
    # Apply Porter stemming
    porter_stem = porter.stem(word)

    # Apply Lancaster stemming (more aggressive)
    lancaster_stem = lancaster.stem(word)

    # Lemmatization using spaCy
    # spaCy processes the word and returns linguistic features
    doc = nlp(word)

    # Extract lemma (base form of the word)
    lemma = doc[0].lemma_

    # Print results in formatted columns
    print(f"{word:<15}{porter_stem:<20}{lancaster_stem:<20}{lemma:<15}")

print("\n" + "=" * 70)

# Sentence-level example to show context-based analysis
sentence = "The children were running and playing happily in the gardens."

# Process entire sentence using spaCy
doc = nlp(sentence)

print("\nSentence Analysis:")
print("Original Sentence:", sentence)

# Print column headers
print("\nToken\t\tLemma\t\tPOS")

# Loop through each token in sentence
for token in doc:
    # token.text → original word
    # token.lemma_ → base form
    # token.pos_ → part-of-speech (noun, verb, etc.)
    print(f"{token.text:<15}{token.lemma_:<15}{token.pos_}")

print("\n" + "=" * 70)

# Final observations printed for understanding
print("\nObservations:")
print("1. Stemming may produce non-dictionary words (e.g., 'running' -> 'run').")
print("2. Lemmatization produces meaningful base forms (e.g., 'better' -> 'good').")
print("3. Lancaster stemmer is more aggressive than Porter stemmer.")
print("4. Lemmatization uses context and part-of-speech for accuracy.")
```
